```python

from surprise import Dataset
from surprise import Reader
from surprise import SVD
from surprise.model_selection import train_test_split
from surprise import accuracy
```

```python
print(dir(Dataset))
```

```python
reader = Reader()
```

```python
surprise_data = Dataset.load_from_df(new_df[['visitorid','itemid','event']] , reader)
```

```python
model = SVD()
```

```python
surprise_data
```

```python
train_X,test_X = train_test_split(surprise_data, test_size = 0.2, random_state = 42)
```

```python
model.fit(train_X)
```

```python
pred = model.test(test_X)
```

```python
print(accuracy.rmse(pred))
```

```python
def recommend_cf(user_id, n=5):
    unique_items = new_df['itemid'].unique()
    recommendations = []

    for item in unique_items:
        prediction = model.predict(user_id, item)
        recommendations.append((item, prediction.est))
    recommendations.sort(key=lambda x: x[1], reverse=True)
    return [item for item, _ in recommendations[:]]
```

```python
user_id = 3 # Example user ID
top_recommendations = recommend_cf(user_id,n = 3)
print(f"Top recommendations for User {user_id}: {top_recommendations}")
```
